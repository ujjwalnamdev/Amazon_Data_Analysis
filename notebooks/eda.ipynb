```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/processed/cleaned_amazon.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['rating'].head()
```

```python
df["rating"] = pd.to_numeric(df["rating"], errors = 'coerce')
```

```python
df
```

```python
#Savings Amount
df["savings_amount"] = df["actual_price"] - df["discounted_price"]
```

```python
df.info()
```

```python
#Review length
df["review_length"] = df["review_content"].astype(str).apply(len)
```

```python
#Clean column
df["rating_count"] = (df["rating_count"]
    .astype(str)
    .str.replace(",", "", regex = False)
)
```

```python
# Convert to numeric
df["rating_count"] = pd.to_numeric(
    df["rating_count"],
    errors="coerce"
)
```

```python
df["popularity_score"] = (
    df["rating"] * df["rating_count"]
)
```

```python
df.info()
```

```python
#Top category
df["category"].value_counts().head(10)
```

```python
# plt.figure(figsize = (10, 5))
plt.figure(figsize=(10,5))
df["category"].value_counts().head(10).plot(kind="bar")
plt.title("Top Categories")
plt.show()
```

```python
plt.figure(figsize=(12,6))

df["category"].value_counts().head(10).plot(
    kind="barh"
)

plt.title("Top Categories")
plt.xlabel("Count")
plt.ylabel("Category")

plt.show()
```

```python
df["main_category"] = df["category"].apply(
    lambda x: x.split("|")[0]
)
```

```python
#simpler graph for 
plt.figure(figsize=(10,5))

df["main_category"].value_counts().head(10).plot(
    kind="barh"
)

plt.title("Top Main Categories")
plt.xlabel("Count")
plt.ylabel("Category")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10,5))

df["main_category"].value_counts().head(10).plot(
    kind="bar"
)

plt.title("Top Main Categories")
plt.xticks(rotation=45)

plt.show()
```

```python
#Rating Distribution
plt.figure(figsize = (8,5))
sns.histplot(df["rating"], bins = 20)
plt.title("Rating Distribution")
plt.show()
```

```python
#Highest Discount Products
df.sort_values(by = "discount_percentage", ascending = False)[
["product_name", "discount_percentage"]
].head(10)
```

```python
#Top rated products
df.sort_values(by = "rating", ascending = False)[
    ["product_name", "rating"]
    ].head(10)
```

```python
#Most popular products
df.sort_values(by = "rating_count", ascending = False)[
    ["product_name", "rating_count"]
].head(10)
```

```python
#correlation Analysis
numeric_df = df.select_dtypes(include = np.number)
plt.figure(figsize = (8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap = "coolwarm")
plt.title("correlation Heatmap")
plt.show()
```

```python
category_counts = df["main_category"].value_counts()
df["category_grouped"] = df["main_category"].apply(
    lambda x: x if category_counts[x] >= 20 else "Other"
)
```

```python
df = df.dropna(subset=["rating", "discount_percentage"])
```

```python
df.to_csv(
    "../data/processed/final_amazon.csv",
    index=False,
    encoding="utf-8-sig"
)
```
